## SQLite Connection with Python¶

```python
#pip install sqlalchemy
```

**What is the use of SQLAlchemy?**<br>

SQLAlchemy is a library that facilitates the communication between Python programs and databases.

```python
from sqlalchemy import create_engine
```

```python
temp_db = create_engine("sqlite:///:memory:")

# sqlite içine geçici-hayalet bir db oluşturduk, on the air
# notebook kapatıldığı an temp_db kayboluyor
```

The above engine creates a **Dialect object** tailored towards sqlite, as well as a **Pool object** which will establish a DB-API connection at local memory when a connection request is first received. 

**NOTE :** DB-API is an acronym for DataBase Application Programming Interface and a library that lets Python connect to the database server.

```python
temp_db.table_names()

# temp_db çağırıyorum. table_names() attribute kullanarak içindeki table lara bakıyorum
# içinin boş olduğunu görüyorum.
```

```python
tables = pd.read_html('https://en.wikipedia.org/wiki/World_population')
tables
```

```python
len(tables)

# 24 tane table çekilmiş
```

```python
tables[2]

# herhangi birisini indexleme ile görüntülüyorum
```

```python
tables[4].to_sql(name='populations', con=temp_db)

# to_sql ile kaydediyoruz
# temp_db database i içerisine populations isimli bir table oluşturduk
```

```python
temp_db.table_names()
```

**How can I open and read a SQL file in python?**

```python
df = pd.read_sql(sql='populations', con=temp_db)
df

# con ile temp_db ile connection yaptık.
# ilgili table ile df olarak read_sql ile okuttuk
```

**How can I make a SQL query in python?**

```python
pd.read_sql_query(sql="SELECT Country, Population FROM populations", con=temp_db)

# temp_db içindeki populations table dan query yoluyla 2 sütun görüntüledik.
# sql de hangi snytax ve hangi query yazılıyorsa burada da yazılabilir
# büyük küçük harf duyarlılığı yok
```

```python
import sqlite3
```

```python
with sqlite3.connect('sample_data4.db') as cnnt:
    df.to_sql('sample1', cnnt)
```

```python
query = 'SELECT Country, Population FROM sample1'
```

```python
with sqlite3.connect('sample_data4.db') as cnnt:
    df_new = pd.read_sql_query(query, cnnt)
```

```python
df_new
```
