# <h1 style="text-align: center;" class="list-group-item list-group-item-action active" data-toggle="list" role="tab" aria-controls="home">Data Visualization - 2</h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

**Use below dataframe for plots**

```python
income = [100, 80, 150, 48, 52, 69, 88]
expense = [30, 100, 100, 20, 75, 50, 28]
index = ['A', 'B', 'C','D', 'E', 'F', 'G']
df = pd.DataFrame({'income': income,'expense': expense}, index=index)
df.head()
```

# Bar plots

## 1. Set darkgrid style from seaborn

```python
# CODE HERE
sns.set_style("darkgrid")
```

## 2. Make a bar plot

```python
# CODE HERE

fig, ax = plt.subplots(figsize=(5, 3))
                               

p = np.arange(len(df.index))
width = 0.20

ax.bar(p - width/2, df.income, width, label="income")
ax.bar(p + width/2, df.expense, width, label="expense")

ax.set_xticks(p)
ax.set_xticklabels(df.index)
plt.xticks(rotation=90)

plt.legend()

plt.show()
```

## 3. Stack the bars

```python
# CODE HERE

fig, ax = plt.subplots(figsize=(5, 3))

ax.bar(df.index, df.income, label="income", width=0.5)
ax.bar(df.index, df.expense, bottom=df.income, label="expense", width=0.5)

plt.xticks(rotation=90)
plt.legend()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))

ax.bar(df.index, df.income, label="income", width=0.5)
ax.bar(df.index, df.expense, label="expense", width=0.5)

plt.xticks(rotation=90)
plt.legend()

plt.show()
```

## 4. Rotate the labels and set figsize

```python
# CODE HERE

fig, ax = plt.subplots(figsize=(6.5, 4.5))

ax.bar(df.index, df.income, label="income", width=0.5)
ax.bar(df.index, df.expense, bottom=df.income, label="expense", width=0.5)

plt.legend()

plt.show()
```

## 5. Replace the labels by "Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"

```python
games = ["Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"]
df["games"] = games
df
```

```python
df = df.set_index("games")
df
```

```python
# CODE HERE

df = df.set_index("games")
df.index.name = None
df
```

## 6. Unstuck the bars, annotate the hight of the bars on top of them

```python
# CODE HERE

fig, ax = plt.subplots(figsize=(6.5, 4.5))
                               

p = np.arange(len(df.index))
width = 0.20

ax.bar(p - width/2, df.income, width, label="income")
ax.bar(p + width/2, df.expense, width, label="expense")

ax.set_xticks(p)
ax.set_xticklabels(df.index)
ax.bar_label(ax.containers[0])

plt.legend()

plt.show()
```

# Histograms

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

## 7. Histogram of the total_bill column

```python
# CODE HERE

figsize=(5, 3)
sns.histplot(x=tips.total_bill)
```

# Lineplots

```python
tips.head()
```

# 8. Plot the avg tip by size  (lineplot)

```python
# CODE HERE
tips.groupby("size").tip.mean()
```

```python
# CODE HERE

av_tip = tips.groupby("size").tip.mean()

figsize=(5,3)

fig, ax = plt.subplots()

ax.plot(av_tip.index, av_tip.values)

ax.set_xlabel("size")

plt.show()
```

## 9.Set the linestyle as "--"

```python
# CODE HERE

av_tip = tips.groupby("size").tip.mean()

figsize=(5,3)

fig, ax = plt.subplots()

ax.plot(av_tip.index, av_tip.values, linestyle="--")

ax.set_xlabel("size")

plt.show()
```

## 10. Make a scatter plot between tip and total_bill
